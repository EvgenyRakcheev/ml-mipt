```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import datasets, cross_validation, metrics, neighbors
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.cross_validation import cross_val_score
from matplotlib.colors import ListedColormap
from pandas import DataFrame
%pylab inline
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
digits_frame = DataFrame(digits.data)
digits_frame['target'] = digits.target
breast_cancer_frame = DataFrame(breast_cancer.data)
breast_cancer_frame.columns = breast_cancer.feature_names
breast_cancer_frame['target'] = breast_cancer.target
```

```python
for dataset in [(digits, "digits"), (breast_cancer, "breast_cancer")]:
    for estimator in [(BernoulliNB(), "BernoulliNB"), (MultinomialNB(), "MultinomialNB"), (GaussianNB(),"GaussianNB")]:
        print(dataset[1], ":", "distribution:", estimator[1], ", accuracy:",
              cross_validation.cross_val_score(estimator[0], dataset[0]['data'], dataset[0]['target']).mean())
```
