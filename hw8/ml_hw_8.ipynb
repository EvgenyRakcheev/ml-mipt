```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(F.T)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot * 100.0 for i in range(n)])
    m = np.argmax(var_i > varRetained * 100)
    print ('{:.2f} % variance retained in {} dimensions'.format(var_i[m], m))

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = V.T
    G = np.dot(F.T, U_reduced)

    return G, U_reduced
```

## Первая часть:

```python
from sklearn import datasets
from sklearn.cluster import DBSCAN, KMeans, MeanShift, AgglomerativeClustering, estimate_bandwidth
from sklearn import metrics
from sklearn.manifold import TSNE
from sklearn.neighbors import kneighbors_graph
from skimage.measure import compare_ssim
import matplotlib.pyplot as plt
```

```python
data = datasets.load_digits()
```

```python
target = data['target']
data = data['data']
```

```python
def dbscan(epsilon = 2, min_samples = 15, plot = 0):    
    model = TSNE()
    tsne_data = TSNE().fit_transform(data)
    db = DBSCAN(eps=epsilon, min_samples=min_samples).fit(tsne_data)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    print('Estimated number of clusters: %d' % n_clusters_)
    print("Adjusted Rand Index: %0.3f"
          % metrics.adjusted_rand_score(target, labels))
    print("Adjusted Mutual Information: %0.3f"
          % metrics.adjusted_mutual_info_score(target, labels))

    if plot == 1:
        unique_labels = set(labels)
        colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
        plt.figure(figsize=(10, 8))
        for k, col in zip(unique_labels, colors):
            if k == -1:
                col = 'k'
            class_member_mask = (labels == k)
            xy = tsne_data[class_member_mask & core_samples_mask]
            plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                     markeredgecolor='k', markersize=14)
            xy = tsne_data[class_member_mask & ~core_samples_mask]
            plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                     markeredgecolor='k', markersize=6)
        plt.title('Estimated number of clusters: %d' % n_clusters_)
        plt.show()
```

## Изменяя eps и min_sample, проверим, при каких значениях метрики дают лучшее значение

```python
for eps in [0.5, 1, 2, 3, 5]:
    for samp in [5, 7, 10, 12, 15, 18, 20]:
        print('epsilon =', eps)
        print('min_samples =', samp)
        dbscan(eps, samp)
```

```python
image = plt.imread('two-lorikeet-birds-2293918.jpg')
np.shape(image)
```

```python
im_X = np.reshape(image, (-1, 3))
```

## Исходная картинка

```python
plt.imshow(image)
plt.show()
```

```python
def plot_new_image(num_clusters = 2, algo = 'KMeans', quantile = 0.2, col = 'mean'):
    if algo == 'KMeans':
        im_X_pred = KMeans(n_clusters=num_clusters).fit_predict(im_X)
    if algo == 'MeanShift':
        bandwidth = estimate_bandwidth(im_X, quantile=quantile, n_samples=500)
        im_X_pred = MeanShift(bandwidth=bandwidth, bin_seeding=True).fit_predict(im_X)
        num_clusters = len(np.unique(im_X_pred))
    if algo == 'aggl':
        kng = kneighbors_graph(im_X, 30, include_self=False)
        im_X_pred = AgglomerativeClustering(connectivity=kng, n_clusters=num_clusters).fit_predict(im_X)
    mean_col = []
    if col == 'mean':
        for cluster in range(num_clusters):
            mean_col.append(np.mean([im_X[i] for i in range(len(im_X)) if im_X_pred[i] == cluster], axis = 0))
    if col == 'median':
        for cluster in range(num_clusters):
            mean_col.append(np.median([im_X[i] for i in range(len(im_X)) if im_X_pred[i] == cluster], axis = 0))
    mean_col = np.array(mean_col) / 255
    im_X_changed = []
    for i in range(len(im_X)):
        for j in range(num_clusters):
            if im_X_pred[i] == j:
                im_X_changed.append(mean_col[j])
    im_X_2 = np.reshape(im_X_changed, (275, 400, 3))
    plt.imshow(im_X_2)
    plt.title('num_clusters = ' + str(num_clusters))
    plt.show()
    print ('SSIM = %s' %compare_ssim(image/255, im_X_2, multichannel=True))
```

## Алгоритм кластеризации KMeans:

```python
for num_clusters in [2, 3, 10, 20]:
    plot_new_image(num_clusters)
```

## Алгоритм кластеризации MeanShift:

```python
for q in [0.2, 0.17, 0.0576, 0.03]:
    plot_new_image(2, 'MeanShift', quantile = q)
```

## Алгоритм кластеризации AgglomerativeClustering:

```python
for num_clusters in [2, 3, 10, 20]:
    plot_new_image(num_clusters, 'aggl')
```
